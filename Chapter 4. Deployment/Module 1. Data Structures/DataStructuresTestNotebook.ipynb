## Time Complexities

When comparing algorithms we use Big-O notation (`O(1)`, `O(n)`, `O(n^2)`) which is useful for providing an approximate measure of how long it takes for an algorithm to complete given some variable input size. 

For example, if the algorithm was to add 2 numbers together, this would take a constant amount of time. This is because even very large numbers can be represented in binary with about 64 bits and adding 64 bits together takes a constant amount of time. We don't know how long it takes but the important thing to remember is that it takes the same amount of time everytime. 

On the other hand, if we're searching for a specfiic number in an unordered list: we need to look through every element in the list, check if it is that element, then return True or keep looking. This means that this searching algorithm takes a time proportional to the size of the list we are searching i.e. `O(n)`. 

If we look through an ordered list, we can go straight to the middle see if the element is greater or smaller then search the correct have of the list, this therefore takes a time proprtional to the logarithm (base 2) of the length of the list i.e.  `O(logn)`. This is because after 1 step, we half the number of elements we need to look through.

# Datastructures

## Lists

Python lists are implemented as dynamic arrays. Picture the list as a single chunk of memory than can be increased or decreased in size. This leads to some interesting characteristics, for example, appending to the end takes a constant amount (represented by `O(1)` of time since the program fills the last block of memory with the new item that is added. However, appending to the front means, every other element of the list needs to be shifted forward one block to make space for the element that is added. This operation therefore takes a time proportional to the number of items in the list i.e. the number of items it has to shift along. This is a linear time operation, which in shorthand is `O(n)`.

## Sets 

Sets are an **unordered** collection of **unique** elements. To find something in a set however takes on average a constant (`O(1)`) amount of time. This is because of the internal implementation of a set. 

## Linked Lists

It doesn't matter where elements of a linked list are stored in memory. Each element simply has a pointer pointing to the next element in the list. So if I want to know the 3rd element in my list, I have to start at the front of the list. The first element tells me how to get to the second, then the second element tells me how to get to the third. Even finding elements takes `O(n)` time. However, adding elements to the front of the list takes `O(1)` time.

## Deques

Double-ended queues (deques) are useful since appending and taking things from both left and right takes `O(1)` time unlike standard lists. This is useful when trying to find a datastructure for a first in first out (FIFO) queue. In that scenario, we need to add to the back and take from the front and ideally both of those operations need to be as efficient as possible. 

## Tuples

Tuples can sometimes be though of like lists as they store a collection of multiple types. Tuples are however immutable (can't change) i.e. `tuple[index] = newElem` isn't allowed. 



## Hashmaps

Hashmaps (also known in python as dictionaries) store a key and a value. We query using the key to receive a value. To get the value given a key takes `O(1)` amount of time. This is the principal reason hashmaps are useful as it doesn't matter how many keys are in the hashmap. 

## Default dictionaries

Python provides a collections library for extra common datastructures used in algorithms. One of these is the `defaultdict`. This datastructure allows us to specify a default value for any new key being added to the dictionary. For example: if we would like to store the index of every occurrence of every digit in a list of digits, we would need to keep the digit as a key and a list of indices as a value. We could use `myDict = defaultdict(list)` which creates an empty list by default. Hence for every digit, instead of checking it's in the dictionary and creating an empty list if it's not we can simply write `myDict[currentDigit].append(currentIndex)`. The append will append to a new list as that is the default value if there isn't another value (i.e. if the `currentDigit` key is not in the list). 


## Counter

This dictionary is another collection but it automatically calculates the count of occurrences of specific types in a sequence. For example, it counts the occurrence of different characters in a string. For example `Counter('gallahad')` returns `Counter({'a': 3, 'l': 2, 'g': 1, 'h': 1, 'd': 1})`.

# Space Complexities

Space complexity aims to represent the proportion of space used by an algorithm in relation to the inputs. For example, if we are inputting a list and copy it during the algorithm, we would say the algorithm uses `O(n)` space (linear). However, if the algorithm only utilised a few constant variables for example the length of the list, the minimum value or the maximum value, then we would say the space complexity is `O(1)`. It is important to realise that every function calls uses some space as well (for storing arguments, return addresses etc.). We call the space used to store function calls the **call stack**. Hence, if we call a function a number of times proportional to the size of an input list, the space complexity would be `O(n)` even without copying a list. If we did copy a list in each of those functions, the space complexity would be `O(n^2)` since there would be a number of functions and a separate list stored for each of those functions. 

# Fibonacci 

Each number in the Fibonacci sequence is the sum of the previous 2 numbers of the sequence. The sequence also starts with 0 and 1. From this defintion we get: 

`0, 1, 1, 2, 3, 5, 8, 13...` (convert)

We can see an immediate algorithm to compute the nth fibonacci number based on its definition. Firstly, if we are looking for the first or second number we would return `0` and `1` respectively according the definition. 

Alternatively, if we are looking for the 5th number for example, we would use recursion and define the answer in terms of another function. We would define the answer as `fib(5) = fib(4) + fib(3)`. 

More generally, `fib(n) = fib(n-1) + fib(n-2)`. 

To investigate the time complexity we look at loops and function calls. In every function there are 2 function calls (assuming the input is >1). To work out both of those functions there will be 2 more function calls each. Hence there will be 4 function calls. Then to work out each of those 4 functions there will be 8 calls and so on. Eventually we will get to `fib(0)` or `fib(1)` which takes a constant amount of time. Therefore we can see the total number of calls is directly proportional to `2^n`. Hence the time complexity is `O(2^n)`.

```
def fib(n):
  if n <= 1:
    return 1
  else:
    return fib(n-1) + fib(n-2)

table = {}

def fib2(n):
  if n in table:
    return table[n]
  if n <= 1:
    return 1
  else:
    table[n] = fib2(n-1) + fib2(n-2)
    return table[n]

# fib(35)
fib2(35)


# [1, 2, 3, 4, 5]  
```

## Merge Sort

Merge sort is the most optimal algorithm in terms of worse case time complexity. It is `O(nlogn)` in the worst case. The algorithm:
  * If the array is of size 1 then it is sorted so return the array.
  * Otherwise, split the array in half and sort both sides using merge sort (recursion). 
  * Merge the 2 sorted left and right arrays. 


Mergesort has `O(n)` space complexity and is slower in practice. This is because it involves copying data into another array and copying is usually slower than having extra comparisons.

```
def merge(left, right):
  leftIndex, rightIndex = 0, 0
  merged = []
  while leftIndex < len(left) and rightIndex < len(right):
    # compare numbers in left and right arrays and append the smaller number
    if left[leftIndex] < right[rightIndex]:
      merged.append(left[leftIndex])
      leftIndex += 1
    else:
      merged.append(right[rightIndex])
      rightIndex += 1
  # append any remaining numbers in 1 of the arrays after reaching the end of the other array. 
  return merged + left[leftIndex:] + right[rightIndex:]

def mergesort(nums):
  if len(nums) <= 1:
    return nums
  else:
    mid = len(nums) // 2
    # sort the left 
    left = mergesort(nums.copy()[:mid])
    # sort the right
    right = mergesort(nums.copy()[mid:])
    # merge the 2 sorted arrays
    return merge(left, right)
```
